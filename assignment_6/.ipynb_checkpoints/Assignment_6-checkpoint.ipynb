## Note: Rerunning kmeans.fit() will cause the labels to change

```python
import numpy as np
import pandas as pd
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib import gridspec
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8, 6
sns.set_style('whitegrid')
```

### Load Data

```python
df = pd.read_csv("Data.txt")
```

## Q1 - Number of Clusters

```python
# Rescaling the data and dropping customer and order features
scaler = MinMaxScaler()
X = scaler.fit_transform(df.drop(['customer', 'order'], axis=1))
```

```python
km_silhouette = []
for i in range(2, 30):
    km = KMeans(n_clusters=i, random_state=0).fit(X)
    preds = km.predict(X)
    silhouette = silhouette_score(X, preds)
    km_silhouette.append(silhouette)
    print("Silhouette score for number of cluster(s) {}: {}".format(i,silhouette))
```

```python
plt.figure(figsize=(12,6))
plt.title("The silhouette coefficient method \nfor determining number of clusters\n",fontsize=16)
plt.plot([i for i in range(2, 30)], km_silhouette)
plt.grid(True)
plt.xlabel("Number of clusters",fontsize=14)
plt.ylabel("Silhouette score",fontsize=15)
plt.xticks([i for i in range(2, 30)],fontsize=14)
plt.yticks(fontsize=15)
plt.show()
```

#### Explanation
8 is the best choice for the number of clusters as it has a high silhouette score while not being too low to be to lose meaning for high dimentional data

## Q2 - Clustering Meathods

### KMeans Model

```python
kmeans = KMeans(n_clusters=8)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
X.shape
```

```python
X[0, :].shape
```

### DBSCAN Model

```python
# DBSCAN Method with non-clustered values removed
dbs = DBSCAN(eps=0.25, min_samples=110, p=2, n_jobs=-1, algorithm='auto')
dbs.fit(X)
y_dbs_raw = dbs.labels_
X_dbs = np.copy(X)
y_dbs = []
indexes = []
for i in range(0, len(y_dbs_raw)):
    if y_dbs_raw[i] > 0:
        y_dbs.append(y_dbs_raw[i])
    else:
        indexes.append(i)
X_dbs = np.delete(X_dbs, indexes, axis=0)
len(set(y_dbs))
```

```python
len(y_dbs)
```

### Graphs

```python
ys = [y_kmeans, y_dbs]
Xs = [X, X_dbs]
title = ["Kmeans", "DBSCAN"]
columns = df.columns[2:]

fig = plt.figure(figsize=(18, 8))
gs = gridspec.GridSpec(1,2)
for i, y in enumerate(ys):
    ax = plt.subplot(gs[i])
    ax.scatter(Xs[i][:, 4], Xs[i][:, 5], c=y, s=50, cmap='tab10')
    if i == 0:
        ax.scatter(kmeans.cluster_centers_[:, 4], kmeans.cluster_centers_[:, 5], c='red', s=200, alpha=0.75);
    ax.set_title(title[i]+"\n%Food vs %Fresh", fontdict={'fontsize':16})
    plt.xlabel("%Food",fontsize=15)
    plt.ylabel("%Fresh",fontsize=15)
```

## Q3 - PCA

```python
pca = PCA(n_components=None)
X_pca = pca.fit(X)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=[i+1 for i in range(len(X_pca.explained_variance_ratio_))],
            y=X_pca.explained_variance_ratio_,
           s=200, alpha=0.75,c='orange',edgecolor='k')
plt.grid(True)
plt.title("Explained variance ratio of the \nfitted principal component vector\n",fontsize=25)
plt.xlabel("Principal components",fontsize=15)
plt.xticks([i+1 for i in range(len(X_pca.explained_variance_ratio_))],fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel("Explained variance ratio",fontsize=15)
plt.show()
```

```python
X_trans = X_pca.transform(X)
X_trans = pd.DataFrame(data=X_trans)
X_trans.head(10)
```

```python
plt.figure(figsize=(14,10))
scatter = plt.scatter(X_trans[0], X_trans[1], c=y_kmeans, edgecolors='k', alpha=0.50, s=50, cmap='tab10')
plt.grid(True)
plt.title("Class separation using first two principal components\n",fontsize=20)
plt.xlabel("Principal component-0",fontsize=15)
plt.ylabel("Principal component-1",fontsize=15)
plt.legend(handles=scatter.legend_elements()[0], labels=labels, loc='upper right')
plt.show()
```
